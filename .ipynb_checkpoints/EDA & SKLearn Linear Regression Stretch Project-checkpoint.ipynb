### About the data:

##### Reference

#### USDA. (n.d.). data.world. Data.World | The Cloud-Native Data Catalog. Retrieved October 7, 2021, from https://data.world/craigkelly/usda-national-nutrient-db/workspace/project-summary?agentid=craigkelly&datasetid=usda-national-nutrient-db

This data is a flattened version of the USDA National Nutrient Database. It is extracted from the MongoDB database created in the project https://github.com/CraigKelly/nndb-import. The original data can be found at http://www.ars.usda.gov/Services/docs.htm?docid=8964

The columns in US RDA are created using data from Dietary Reference Intakes: The Essential Guide to Nutrient Requirements available from the National Academies Press at http://www.nap.edu/catalog/11537.html



### Data Description:

Each record is for 100 grams.

The columns are mostly self-explanatory. The nutrient columns end with the units, so:

- Nutrient_g is in grams
- Nutrient_mg is in milligrams
- Nutrient_mcg is in micrograms
- Nutrient_USRDA is in percentage of US Recommended Daily Allows (e.g. 0.50 is 50%)
- Note that not every available RDA value in the text was used. For instance, the US RDA for Iron varies significantly by age and sex, so I deemed it easier to just leave out RDA information for Iron.

## Exploratory Data Analysis

Going forward we will be addressing a prediction for Energy_Kcal. We want to predict calories given certain features. I would like to understand calories a bit more.I'll setup some central limit metrics just to get a better understanding.

```python
usda['Energy_kcal'].describe()
```

Quite a high standard deviation. I wonder if median is any better. Given the sample size is an even count number of 8618 it will equal the 50th percentile.

```python
print('Median: ', usda['Energy_kcal'].median())
```

I would like to visualize this, reviewing the skewness and kurtosis to understand the standard deviation a bit better. What is our distribution?

```python
print(f'Skewness: {usda["Energy_kcal"].skew()}')
print(f'Kurtosis: {usda["Energy_kcal"].kurt()}')

sns.distplot(usda['Energy_kcal']);
```

This skewness number tells us that it is not a normally distributed sample. However, we did not expect this for Calories given the standard deviation was quite high and the mean was higher than the median. Given that the skewness is 1.168, this confirms a right tailed distribution. In the graph we see many outliers within this tail. We also know that we cannot have zero calories, so we would have expected at least a trail-off to zero from the mean/median that we do not get.

In kurtosis a positive number indicates a heavy-tailed distribution. As we see in the graph above this falls in line with extreme outcomes, in this case high calorie examples.

# Linear Regression Model

Using Sklearn we will create a linear regression model for the categorical variable FoodGroup. Beginning with getting dummy variables for FoodGroup. Our goal is to predict calories. We will need to define X and Y. 

```python
usda = pd.get_dummies(usda, columns=['FoodGroup'], drop_first=True)
usda.shape
```

```python
usda
```

We then need to have a training and testing set. So I will use sklearn's built in model selection function.

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(usda, test_size=0.2)
```

We create the X_train which is all data less the Y variable we seek to predict

```python
X_train = train[1:]
X_train.shape
```

```python
# X_train.head()
```

We then create the y_train which is our predicted variable.

```python
y_train = train['Energy_kcal']
y_train.shape
```

```python
y_train.head()
```

There appears to be a mismatch in the records for some reason. We are not matching the rows by 1. So I will drop this first row in the y_train dataset to match. 

```python
y_train = y_train.drop(y_train.index[0])
y_train.shape
```

```python
y_train.head()
```

We can now create a linear model from the train datasets. So I will now import the linear model from sklearn.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

I will now plug in the test data that we create earlier from the sklearn train_test_split function. This will be our predicted.

```python
X_test = test[1:]
X_test.head()
```

Now we create a y_test variable similarly to above. This is our learning stage of machine learning.

```python
y_test = test['Energy_kcal']
y_test.head()
```

Similar to before we have an additional row. This will be dropped to match.

```python
y_test = y_test.drop(y_test.index[0])
y_test.head()
```

The next step will test our model using the `predict()` function. We can then evalutate.

```python
pred = reg.predict(X_train)
```

```python
print(f'Mean squared error: {mean_squared_error(y_train, pred)}')
```

```python
print(f'r2 score (1 is perfect): {r2_score(y_train, pred)}')
```

# Potential next step....Testing Normality

```python
from scipy.stats import norm
import scipy.stats as stats
```

```python
sns.distplot(usda['Energy_kcal'], fit=norm);
fig = plt.figure()
res = stats.probplot(usda['Energy_kcal'], plot=plt)
```

Not very normal I would say! I will attempt to transform this a bit.

I will use the sklearn preprocessing function. It involves a MinMaxScaler. Min-max approach (often called normalization) rescales the feature to a hard and fast range of [0,1] by subtracting the minimum value of the feature then dividing by the range.

```python

from sklearn.preprocessing import MinMaxScaler
import numpy as np
  
# copy the data
df_sklearn = usda.copy()
  
# apply normalization techniques
column = 'Energy_kcal'
df_sklearn[column] = MinMaxScaler().fit_transform(np.array(df_sklearn[column]).reshape(-1,1))
  
# view normalized data  
display(df_sklearn)
```

Let's see how this changes the visualization prior to normalization.

```python
sns.distplot(df_sklearn['Energy_kcal'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_sklearn['Energy_kcal'], plot=plt)
```

No change it appears at all. Maybe log will be better.

```python
usda['Energy_kcal'] = np.log(train['Energy_kcal'])
```

```python
usda.head()
```

```python
usda['Energy_kcal'].describe()
```

After making this log transformation we see that we now have infinite values. This will not help us given that we need some kind of linear relationship. So I will replace them with NaN and then see what the null percentage is to determine if they need to be dropped.

```python
usda.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
(usda.isnull().sum())*100/len(usda)
```

Roughly 20%. I believe this is worth the examination of this dataset so we will drop these null values.

```python
usda = usda.dropna()
```

```python
(usda.isnull().sum())*100/len(usda)
```

```python
sns.distplot(usda['Energy_kcal'], fit=norm);
fig = plt.figure()
res = stats.probplot(usda['Energy_kcal'], plot=plt)
```

# Join of Consumer Price Index, EIA, & Inflation Datasets

The following notebook is to create the final dataset to be used for the remainder of this project. I am now joining the cleaned Consumer Price Index, EIA Api, and Inflation datasets. This will be exported to its final CSV at the end.

```python
# libraries to be used to complete work
import pandas as pd
# import numpy as np
# import statsmodels.api as sm
# import pylab
# import matplotlib.pyplot as plt
# import scipy.stats as stats
```

Reading in the prior cleaned datasets of CPI and EIA

```python
cpi = pd.read_csv(r'C:\Users\stanl\Regis Practicum_1\CPI_cleaned.csv')
eia = pd.read_csv(r'C:\Users\stanl\Regis Practicum_1\EIA.csv')
```

```python
# Preview of EIA
eia.head()
```

```python
# preview of CPI
cpi.head()
```

```python
# Dropping the column that represented a prior index when index reset was used.
cpi.drop('Unnamed: 0', axis=1, inplace=True)
eia.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
eia.sort_values('Date')
```

The merge is ready to be done on the Date of each dataset. 

```python
eia_cpi = pd.merge(eia, cpi, on='Date', how='left')
```

```python
# Confirm the merge
eia_cpi.head()
```

Quick check of null values for later.

```python
sort = eia_cpi.isnull().sum()/len(eia_cpi)*100
```

```python
sort.sort_values(ascending=False)
```

Load, drop unnamed and merge Inflation dataset which represents historical inflation rate.

```python
inf = pd.read_csv('historic_inflation.csv')
inf.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
eia_cpi_inf = pd.merge(eia_cpi, inf, on='Date', how='left')
```
